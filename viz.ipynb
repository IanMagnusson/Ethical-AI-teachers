```python
import pandas as pd

df_olmo = pd.read_csv('results/metrics/allenai--OLMo-2-1124-7B-Instruct_scores.csv')
df_olmo
```

```python
df_qwen = pd.read_csv('results/metrics/Qwen--Qwen2.5-7B-Instruct-1M_scores.csv')
df_qwen
```

```python
# Extract relevant data for plotting
scenarios = [
    "mbpp_just_feedback_gpt-3.5-turbo",
    "mbpp_retrieval_as_feedback_gpt-3.5-turbo_no_cheat_no_question",
    "mbpp_just_feedback_rag_teacher_gpt-3.5-turbo_no_cheat_no_question",
    "mbpp_just_feedback_gpt-3.5-turbo_no_cheat_no_question",
    "mbpp_with_feedback_gpt-3.5-turbo_no_cheat_no_question",
    "mbpp_just_feedback_gpt-3.5-turbo",
    "mbpp_with_rag_teacher_feedback_gpt-3.5-turbo_no_cheat_no_question",
    "baseline"
]

# Filter data for the selected scenarios
olmo_filtered = df_olmo[df_olmo["setup"].isin(scenarios)]
qwen_filtered = df_qwen[df_qwen["setup"].isin(scenarios)]

# Models and corresponding accuracy scores
models = ["OLMo (Base)", "OLMo (Plus)", "Qwen (Base)", "Qwen (Plus)"]
data = {
    "Before Feedback": [
        olmo_filtered[olmo_filtered["setup"] == scenarios[-1]]["base_accuracy"].values[0],
        olmo_filtered[olmo_filtered["setup"] == scenarios[-1]]["plus_accuracy"].values[0],
        qwen_filtered[qwen_filtered["setup"] == scenarios[-1]]["base_accuracy"].values[0],
        qwen_filtered[qwen_filtered["setup"] == scenarios[-1]]["plus_accuracy"].values[0],
    ],
    "Just Feedback (cheat)": [
        olmo_filtered[olmo_filtered["setup"] == scenarios[0]]["base_accuracy"].values[0],
        olmo_filtered[olmo_filtered["setup"] == scenarios[0]]["plus_accuracy"].values[0],
        qwen_filtered[qwen_filtered["setup"] == scenarios[0]]["base_accuracy"].values[0],
        qwen_filtered[qwen_filtered["setup"] == scenarios[0]]["plus_accuracy"].values[0],
    ],
    "Retrieval Feedback": [
        olmo_filtered[olmo_filtered["setup"] == scenarios[1]]["base_accuracy"].values[0],
        olmo_filtered[olmo_filtered["setup"] == scenarios[1]]["plus_accuracy"].values[0],
        qwen_filtered[qwen_filtered["setup"] == scenarios[1]]["base_accuracy"].values[0],
        qwen_filtered[qwen_filtered["setup"] == scenarios[1]]["plus_accuracy"].values[0],
    ],
    "Just Feedback (RAG, No Cheat No Question)": [
        olmo_filtered[olmo_filtered["setup"] == scenarios[2]]["base_accuracy"].values[0],
        olmo_filtered[olmo_filtered["setup"] == scenarios[2]]["plus_accuracy"].values[0],
        qwen_filtered[qwen_filtered["setup"] == scenarios[2]]["base_accuracy"].values[0],
        qwen_filtered[qwen_filtered["setup"] == scenarios[2]]["plus_accuracy"].values[0],
    ],
    "Just Feedback (No Cheat No Question)": [
        olmo_filtered[olmo_filtered["setup"] == scenarios[3]]["base_accuracy"].values[0],
        olmo_filtered[olmo_filtered["setup"] == scenarios[3]]["plus_accuracy"].values[0],
        qwen_filtered[qwen_filtered["setup"] == scenarios[3]]["base_accuracy"].values[0],
        qwen_filtered[qwen_filtered["setup"] == scenarios[3]]["plus_accuracy"].values[0],
    ],
    "After Feedback (No Cheat No Question)": [
        olmo_filtered[olmo_filtered["setup"] == scenarios[4]]["base_accuracy"].values[0],
        olmo_filtered[olmo_filtered["setup"] == scenarios[4]]["plus_accuracy"].values[0],
        qwen_filtered[qwen_filtered["setup"] == scenarios[4]]["base_accuracy"].values[0],
        qwen_filtered[qwen_filtered["setup"] == scenarios[4]]["plus_accuracy"].values[0],
    ],
    "After Feedback (cheat)": [
        olmo_filtered[olmo_filtered["setup"] == scenarios[5]]["base_accuracy"].values[0],
        olmo_filtered[olmo_filtered["setup"] == scenarios[5]]["plus_accuracy"].values[0],
        qwen_filtered[qwen_filtered["setup"] == scenarios[5]]["base_accuracy"].values[0],
        qwen_filtered[qwen_filtered["setup"] == scenarios[5]]["plus_accuracy"].values[0],
    ],  
    "With RAG Teacher Feedback (No Cheat No Question)": [
        olmo_filtered[olmo_filtered["setup"] == scenarios[6]]["base_accuracy"].values[0],
        olmo_filtered[olmo_filtered["setup"] == scenarios[6]]["plus_accuracy"].values[0],
        qwen_filtered[qwen_filtered["setup"] == scenarios[6]]["base_accuracy"].values[0],
        qwen_filtered[qwen_filtered["setup"] == scenarios[6]]["plus_accuracy"].values[0],
    ],
}

# Convert data to DataFrame for plotting
df_plot = pd.DataFrame(data, index=models)

# Plot the bar chart
fig, ax = plt.subplots(figsize=(16, 8))
width = 0.1  # Adjust width for multiple bars
x = np.arange(len(models))

# Plot each scenario as a separate bar set
for i, (label, values) in enumerate(df_plot.items()):
    ax.bar(x + (i - 3) * width, values, width, label=label, alpha=0.7)

# Labels and Titles
ax.set_xlabel("Models")
ax.set_ylabel("Pass@1 Score")
ax.set_title("Performance Across Different Feedback and Retrieval Scenarios")
ax.set_xticks(x)
ax.set_xticklabels(models, rotation=15)
ax.legend()

# Display values on bars
for i, (label, values) in enumerate(df_plot.items()):
    for j, value in enumerate(values):
        ax.annotate(f"{value:.3f}",
                    xy=(x[j] + (i - 3) * width, value),
                    xytext=(0, 3),  # Offset text slightly above bars
                    textcoords="offset points",
                    ha='center', va='bottom')

plt.show()
```

```python
# save the plot into png and pdf
fig.savefig('figures/scores.pdf')
fig.savefig('figures/scores.png')
```
